## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
```

```python
url = "https://en.wikipedia.org/wiki/Universal_Pictures"

response = requests.get(url)
soup = bs4.BeautifulSoup(response.content, 'html.parser')
text_content = soup.get_text()

with open('page_text_no_tags.txt', 'w', encoding='utf-8') as file:
    file.write(text_content)

text_content
```

```python
image_tag = soup.find('img')

image_url = image_tag['src']

if image_url.startswith('//'):
    image_url = 'https:' + image_url
elif image_url.startswith('/'):
    image_url = url + image_url

image_response = requests.get(image_url)

with open('downloaded_image.jpg', 'wb') as file:
    file.write(image_response.content)

print(f"Image downloaded and saved as 'downloaded_image.jpg'")
```
